```python
import json
from transformers import pipeline
```

# Transformer pipeline


```python
qa_pipeline = pipeline(
    "question-answering",
    model="mrm8488/spanbert-finetuned-squadv1",
    tokenizer="mrm8488/spanbert-finetuned-squadv1"
)
```

# Load data



```python
#Input file
f = open(path+'dev-v1.1.json', 'r')
dev = json.load(f)
f.close()
```

# Get answer from question and context

```python
count = 0
Scores = []

try:
  for topic in dev['data']:
    for para in topic['paragraphs']:
      cont = para['context']
      for q in para['qas']:
        ques = q['question']
        response = qa_pipeline({
            'context': cont,
            'question': ques
        })
        ans = response['answer']
        score = response['score']
        Scores.append(score)
        count += 1
        print(count)
        #Display output
        # print(str(count) + 'Q: ' + ques)
        # print(str(count) + 'A: ' + ans)
        # print(str(count) + 'S: ' + str(score))
        # print()
        if(count == 1000):
          raise StopIteration
except StopIteration:
  pass
      
```

```python
Avg = sum(Scores) / count
Min = min(Scores)
Max = max(Scores)
print("Avg: ", Avg)
print("Min: ", Min)
print("Max: ", Max)
```
